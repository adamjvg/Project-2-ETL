```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
pokemon_file1 = "pokemon.csv"
poke_df1 = pd.read_csv(pokemon_file1)
poke_df1.head()
```

```python
pokemon_file2 = "Pokemon-2.csv"
poke_df2 = pd.read_csv(pokemon_file2)
poke_df2.head()
```

```python
poke_df1["name"]
```

```python
mergedf= pd.merge(poke_df1, poke_df2, left_on='name', right_on='Name', how='left')
mergedf.head()
mergedf.to_csv('merge_df.csv')
```

```python
mergedf.drop(columns =['is_legendary', 
                       'type1', 
                       'type2', 
                       'pokedex_number', 
                       'sp_attack', 
                       'sp_defense',
                       'speed',
                       'type1',
                       'type2', 
                       'generation',
                       'base_egg_steps' ,
                       'base_happiness' , 
                       'capture_rate', 
                       'percentage_male', inplace = True)
mergedf.head()
```

```python
mergedf.set_index('Name')
mergedf
```

```python
mergedf2 = mergedf.T.drop_duplicates().T
```

```python
mergedf2
```

```python
gen6 = mergedf2[mergedf2.generation != 7]
gen6
```

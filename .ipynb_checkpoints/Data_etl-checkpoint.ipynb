```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np

#set all columns to be displayed
pd.set_option('display.max_columns', None)
```

### Find and store CSV into DataFrame from our first source 

```python
# First source of data 
# Data web source is from https://www.kaggle.com/abcsds/pokemon?select=Pokemon.csv
#import CSV 1

csv_file = "Resources/pokemon.csv"
poke_df1 = pd.read_csv(csv_file)
poke_df1.head()
```

```python
#Check the shape of CSV 1
poke_df1.shape
```

```python
#Agrouped by generation for understand the data and displace it 
poke_df1_group = poke_df1.groupby(['generation','name']).first()
poke_df1_group
```

### Find and store CSV into DataFrame from our second source

```python
# Second source of data 
# Data web source is from https://data.world/ljvmiranda921/pokemon-go-dataset/workspace/file?filename=pkmn-go.csv
#Load CSV 2

csv_file = "Resources/pkmn-go.csv"
poke_df2 = pd.read_csv(csv_file)
poke_df2.head()
```

```python
# Sorting data in alphabetic orded to compare with previous data 1
poke_df2.sort_values(by="name", ascending=True)
```

```python
#check the shape of CSV2
poke_df2.shape
```

## Merge both data source

```python
#merge the two CSV, using right joint
mergedf= pd.merge(poke_df1, poke_df2, left_on='name', right_on='name', how='right')
mergedf.columns
```

```python
#Check the shape of the newly created CSV
mergedf.shape
```

## Clean up our merged data and remove unnecesary information


```python

#drop columns that are not relevant : 'percentage_male', 'url', 'base_egg_steps', 'base_happiness'
coldropdf = mergedf.drop(['percentage_male', 'url', 'base_egg_steps', 'base_happiness', 'type1', 'type2', 'capture_rate_x'], axis = 1)
coldropdf.shape
```

```python
first_column = coldropdf.pop('name')
coldropdf.insert(0, 'name', first_column)
coldropdf.head()
```

```python
#inspect df
coldropdf.info()
```

```python
coldropdf['against_bug'].value_counts()
```

```python
#get pandas to be able to display max of 500 rows per view
pd.set_option('display.max_rows', 500)
```

```python
#get pandas to display all rows that contain NaN
coldropdf[coldropdf.isna().any(axis=1)]
```

```python
#For some Pokemon types, they do not have Type 2 & secondary form, therefore having NaN values in that column is acceptable. 
#Some data are also missing for the weight columns for some Pokemon, however as some analysis might be done without
#requiring said column to have a value, we will ignore NaN values in weight column.
#However, several rows have completely NaN values across all columns - we will remove these (row 28 & 31)
```

```python
# Final clean dataframe
cleandf = coldropdf.drop(coldropdf.index[[28, 31]])
cleandf
```

### Create connection MONGO

```python

# import dependecy
import pymongo
import pandas as pd
```

```python
# Connect to MongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.Pokemons2_DB

try: 
    conn = client
    print("Connected successfully!!!") 
except:   
    print("Could not connect to MongoDB") 

data = cleandf.to_dict(orient='records') 

try: 
    db = client['Pokemon2_db']
    collection = db['Pokemon2']
    # Insert collection
    collection.insert_many(data)
    print("Pokemon Data inserted successfully!!")
except: 
    print("Pokemon Data upload failed")
```

### Create new data with select columns

```python
#Load CSV 2
csv_file = "Resources/pkmn-go.csv"
poke_df2 = pd.read_csv(csv_file)
poke_df2.head(200)
```

```python
#check the shape of CSV2
poke_df2.shape
```

```python
#merge the two CSV, using right joint
mergedf= pd.merge(poke_df1, poke_df2, left_on='name', right_on='name', how='right')
mergedf.columns
```

```python
#Check the shape of the newly created CSV
mergedf.shape
```

```python
#drop columns that are not relevant : 'percentage_male', 'url', 'base_egg_steps', 'base_happiness'
coldropdf = mergedf.drop(['percentage_male', 'url', 'base_egg_steps', 'base_happiness'], axis = 1)
coldropdf.shape
```

```python
first_column = coldropdf.pop('name')
coldropdf.insert(0, 'name', first_column)
coldropdf.head(500)
```
